```python
import numpy as np
import matplotlib.pyplot as plt



X= np.array([2.1, 2.5, 3.6, 4.0])
print(np.mean(X))

Y = np.array([8, 10, 12, 14])
print(np.mean(Y))

covtable=np.cov(X,Y)


print(covtable)

plt.plot(X,Y)
plt.show()
```
