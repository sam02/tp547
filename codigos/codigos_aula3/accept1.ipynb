```python
import numpy as np
import matplotlib.pyplot as plt

n=10000
xfunction=np.array([])

for i in range(n):
    x1=np.random.uniform(0,1)
    x2=np.random.uniform(0,1)

    while (x2 >=(256/27)*x1*pow((1-x1),3)):
        x1=np.random.uniform(0,1)
        x2=np.random.uniform(0,1)
    xfunction = np.append(xfunction, x1)

print(xfunction)
X=np.arange(0, 1, 0.1)
fx=20*X*pow((1-X),3)
plt.plot(X,fx)
plt.hist(xfunction,bins=100,density=True)
plt.show()

```
