```python
import numpy as np
import matplotlib.pyplot as plt
N = 100000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
yexp = -np.log(1 - y * (1 - np.exp(-1)))
#calcular a integral
integral = np.sum((np.exp(-yexp**2)) / ((np.exp(-yexp)) / (1 - np.exp(-1)))) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico
x = np.linspace(0, 1, N)
curve_y = np.exp(-x**2)
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Integral = {integral:.4f}', fontsize=12, color='red')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
