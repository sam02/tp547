```python
import numpy as np
import matplotlib.pyplot as plt
value=5
count=0
p = 0.5;
N = 400;
Media1 = np.array([])
squaremean1 = np.array([])
for j in range(1,10000):
    x = np.zeros(N+1);
    for i in range(1,N):
        u = np.random.uniform();
        if (u <= p):
            x[i + 1] = x[i] + 1;
        else:
             x[i + 1] = x[i] - 1;



    Media1 = np.append(Media1,x[N]);
    squaremean1 = np.append(squaremean1,x[N]**2)


deslmedio = np.mean(Media1);
deslsquare = np.mean(squaremean1);
desviopadrao = np.sqrt(deslsquare - deslmedio**2);
print(deslmedio)
print(desviopadrao)




for dist1 in Media1:
    if dist1==value:
        count=count+1
prob=count/10000
print("a probabilidade e",prob)
```
