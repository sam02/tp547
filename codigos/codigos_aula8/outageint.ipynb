```python
import numpy as np
import matplotlib.pyplot as plt


N=100000 #numero de amostras
d=1 #distancia
dint=0.5 #distância do interferente
alpha1=4 #pathloss exponent
media=d**(-alpha1) #pathloss medio
media2=dint**(-alpha1) #pathloss medio



R=1 #taxa esperada

OP1=[]
N0=1 #densidade espectral da potência do Ruido
for j in np.arange(0,21,1): #Diferentes valores de potência em dB
    pint=10
    plin=10**(j/10) #db para linear
    h =np.random.exponential(1/media,N) #|h|^2
    h2 =np.random.exponential(1/media2,N) #|hint|^2
    Prob1 = np.sum(np.log2(1 + (h*plin /(h2*pint+N0))) < R) / N

    OP1=np.append(OP1,Prob1)





Pdb=np.arange(0,21,1)
plt.semilogy(Pdb,OP1)

plt.grid()
plt.axis([0,20,1e-3,1])
plt.xlabel('P(dB)')
plt.ylabel('Outage probability')
plt.figlegend(['Analítico','simulado'])
plt.show()
```
