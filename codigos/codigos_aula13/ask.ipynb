```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,30,2000)


signal = np.sin(2*np.pi*2.0*t)
data=(np.random.rand(10)>0.5)
data_sequence = np.repeat(data,200)

ask=signal*data_sequence
plt.figure(1)
plt.plot(t, signal)
plt.figure(2)
plt.plot(t,data_sequence)
plt.figure(3)
plt.plot(t,ask)
plt.show()
```
