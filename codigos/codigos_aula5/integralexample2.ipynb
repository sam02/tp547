```python
import numpy as np
import matplotlib.pyplot as plt

N=1000000
y=np.random.uniform(0,1,N)
integral=5*sum(pow((5*y+3),2)*np.exp((-(5*y+3)**2)/2))/N
print(integral)
```

```python
# Define the function
def f(x):
  return x**2*np.exp((-x**2)/2)

# Create the x-axis values
x = np.linspace(3, 8, 100)


# Plot the function
plt.plot(x, f(x))

# Fill the area under the curve
plt.fill_between(x, f(x), color='lightblue')

# Add the label with the area value
plt.text(4, 0.01,f'Area = {integral:.3f}', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of x**2*exp((-x**2)/2) with Area Label')

# Show the plot
plt.show()
```
