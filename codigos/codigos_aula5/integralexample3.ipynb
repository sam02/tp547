```python
import numpy as np
import matplotlib.pyplot as plt

N=1000000
y=np.random.uniform(0,1,N)
integral=1*sum(np.exp((-(1/y-1)/2))/y**2)/N
print(integral)
```

```python
# Define the function
def f(x):
  return np.exp((-x)/2)

# Create the x-axis values
x = np.linspace(0, 100, 100)


# Plot the function
plt.plot(x, f(x))

# Fill the area under the curve
plt.fill_between(x, f(x), color='lightblue')

# Add the label with the area value
plt.text(0.5, 0.3,f'Area = {integral:.3f}', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of exp((-x)/2) with Area Label')

# Show the plot
plt.show()
```
